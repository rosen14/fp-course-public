## Advent of Code 2022 - Day 2


This exercise corresponds to the [day 2 of Advent of Code 2022](https://adventofcode.com/2022/day/2):

The Elves begin to set up camp on the beach. To decide which tent will be closest to the refreshment storage, a giant rock-paper-scissors tournament is already underway.

Rock Paper Scissors is a game between two players. Each game contains many rounds; in each round, players simultaneously choose rock-paper-scissors using a hand shape. Then, a winner is selected for that round: Rock beats Scissors, Scissors beats Paper, and Paper beats Rock. If both players choose the same shape, the round ends in a draw.

Grateful for your help [in the previous exercise](en/63_Exercises.ipynb), an Elf gives you an encrypted strategy guide (your puzzle entry) that will surely help you win. "The first column is what your opponent is going to play: A for Rock, B for Paper, and C for Scissors. The second column..."

Suddenly, the Elf is called to help with someone's shop.

Suppose then that the second column should be the play in response: X for Rock, Y for Paper, and Z for Scissors. Winning every time would be suspicious, so the answers must have been carefully chosen.

The winner of the entire tournament is the player with the highest score. His total score is the sum of his scores for each round. The score for a single round is the score for the shape you selected (1 for Rock, 2 for Paper, and 3 for Scissors) plus the score for the outcome of the round (0 if you lost, 3 if the round was a tie, and 6 if you win).

Since you can't be sure if the Elf is trying to help or not, you have to calculate the score you would get if you followed the strategy guide.

For example, given the following strategy guide:

to and
B X
C Z

This strategy guide predicts and recommends the following:

- In the first round, your opponent will choose Rock (A), so Paper (Y) must be chosen. This ends in a win for one with a score of 8 (2 because you chose Paper + 6 because you won).
- In the second round, your opponent will choose Paper (B), so you must choose Stone (X). This ends in a self defeat with a score of 1 (1 + 0).
- The third round is a draw with both players choosing Scissors, giving you a score of 3 + 3 = 6.
In this example, if you were to follow the strategy guide, you would get a total score of 15 (8 + 1 + 6).

What would your total score be if everything goes exactly according to your strategy guide?

After a while, the elf finishes helping with the shop and sneaks back to you. "Anyway, the second column says how the round should end: X means you should lose, Y means you should end the round in a tie, and Z means you should win. Good luck!"

The total score is still calculated in the same way, but now you have to figure out which way to choose so that the round ends as indicated.

The previous example is now like this:

- In the first round, your opponent will choose Stone (A), and the round needs to end in a draw (Y), so you also choose Stone. This gives you a score of 1 + 3 = 4.
- In the second round, your opponent will choose Paper (B) and you choose Rock, so he loses (X) with a score of 1 + 0 = 1.
- In the third round, your opponent's Scissors must be beaten with Rock by a score of 1 + 6 = 7.
Now with the top secret strategy guide correctly cracked, this would give you a total score of 12.

Following the goblin's new instructions for the second column, what would your total score be if everything goes exactly according to your strategy guide?
