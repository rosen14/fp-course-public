### Advent of Code 2022, day

[Advent of Code](https://adventofcode.com/2022/about) is a series of programming exercises that are published before Christmas each year, hence its
name of _Advent_. Prepared by [Eric Wastl](https://twitter.com/ericwastl), its protagonists are generally the elves who help Santa Claus with
preparation of gifts. This exercise corresponds to the [day 1 of 2022](https://adventofcode.com/2022/day/1):

The Elves have to go to the forest to look for food for the deer, which is left across the sea, and they are kind enough to invite him on this mission. The forest is overgrown and difficult to navigate by vehicle or access from the air; whereupon the Elven expedition traditionally goes on foot. As their ships come closer to land, the Elves begin to take inventory of their supplies. An important consideration is food, in particular, the number of calories each Elf takes in (which will be their _input_ from exercise).

The Elves take turns writing down the number of calories contained in the different meals, snacks, portions, etc. they have brought with them, one item per line. Each Elf separates its own inventory from the previous Elf's inventory (if any) with a blank line.

For example, suppose the Elves finish writing the calories for their items and end up with the following list:

```
1000
2000
3000

4000

5000
6000

7000
8000
9000

10000
```

This list represents the calories in the food that five Elves eat:

- The first Elf carries food with 1000, 2000 and 3000 calories, a total of 6000 calories.
- The second elf carries a food with 4000 calories.
- The third Elf carries food with 5,000 and 6,000 calories, a total of 11,000 calories.
- The fourth Elf carries food with 7,000, 8,000 and 9,000 calories, a total of 24,000 calories.
- The fifth Elf carries a food with 10,000 calories.

In case the Elves are hungry and need extra snacks, they need to know how many calories the Elf with the most calories is carrying. In the example above, this is 24000 (the fourth Elf).

- Find the leprechaun that has the most calories. How many total calories does that elf have?

By the time one finds the answer to the Elves' question, they have already figured out that the Elf who carries the most food calories might run out of food. To avoid this unacceptable situation, the Elves would like to know the total calories carried by the three Elves who carry the most calories. That way, even if one of those Elves runs out of snacks, they still have two backups.

In the example above, the top three Elves are the fourth Elf (with 24,000 calories), then the third Elf (with 11,000 calories), then the fifth Elf (with 10,000 calories). The sum of the calories that these three goblins carry is 45,000.

- Find the top three Elves that carry the most calories. How many calories do those goblins have in total?

The complete file with the caloric inventory of the Elves is located at [here](../data/AoC/2022/1/input).

### Separating into syllables

One of the benefits of the Spanish language, or Castilian as we call it in these latitudes, is
that there are clear rules for [perform the separation of a word into syllables](https://unibetas.com/como-separar-silabas). Syllables are the way letters are grouped when speaking.
In Spanish, the fundamental rule is that _in every syllable there must be at least one vowel_. Starting
Apart from this, there is a set of rules that are used to determine hyphenation.
Some of them are:

- Consonants at the beginning of a word: All consonants found at the beginning of a word must go without exception with the vowel that follows them.

- A consonant located between two vowels: If there is a consonant between two vowels in the word, it must accompany the one that follows it:

- Two consonants located between two vowels: In this case, the separation of syllables will depend on the specific group of consonants. If they are “pr”, “br”, “dr”, “cr”, “fr”, “gr”, “kr”, “tr”, “fl”, “pl”, “gl”, “kl”, “cl” or “bl”, are kept together at all times and are linked to the vowel that follows them, as in Alegría: a – le – grí – a.

It is proposed to think (not program, unless you so wish) how to approach the programming of a
code that given a word of the Spanish language, separates it into syllables.
