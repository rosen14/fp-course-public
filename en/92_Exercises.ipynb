### Linear Interpolation

We have been tasked to write a function that given two lists `x` and `y` of floating point numbers,
and given a specific value `x0`, let us find the corresponding value of `y0` by linear interpolation.

Due to the AI ​​explosion, we decided to ask [ChatGPT](https://openai.com/blog/chatgpt) to write us the code
in F#, and we will focus on performing unit tests to prove that the generated code works correctly.

So:

- Create a solution to organize the code. In this solution believe

- a project containing the library of functions to be used in linear interpolation.
- an NUnit test project to test them.

- Get the ChatGPT code, (or write it yourself...)

- Write the tests to test the code, under normal conditions:

- `x` and `y` have the same size `n`.
- `x` has all different elements, and in increasing order.
- `x0` is between the values ​​`x[0]` and `x[n-1]`.

- Now modify the code to return a `Result<float,string>` type, so that an error is thrown if

- `x` and `y` have different sizes.
- `x` has equal elements

and return the interpolated `Ok y0` value if the validations are successful.

- Modify the tests to contemplate the previous cases.


#### Extrapolation

Now they tell us that the code has to contemplate the case of extrapolation,
which happens when `x0 < x[0]` or `x0 > x[n-1]` such that:

- If `x0 < x[0]`, return the value `y0 = y[0]`.
- Return a new error if `x0 > x[n-1]`.

Modify the library and the tests according to this new requirement.

> Note that it is not necessary to create a project that _uses_ the library, just make the tests...
